```python
#Logica de programação

# Entender o que está sendo proposto

#Percorrer todos os arquivos da pasta de dados

import os
import pandas as pd
import plotly.express as px

tabela_faturamento = tabela_faturamento[["faturamento"]].sort_values(by='faturamento', ascending=False

lista_arquivo = os.listdir('/content/drive/MyDrive/Python/Vendas')
tabela_total = pd.DataFrame()

#Importar a base de dados

for arquivo in lista_arquivo:
  if 'Vendas' in arquivo: 
    tabela = pd.read_csv(f'/content/drive/MyDrive/Python/Vendas/{arquivo}')
    tabela_total = tabela_total.append(tabela)


#Compilar as bases de dados

display(tabela_total)
```

```python
#Calcular o produto que mais vendeu

tabela_produtos = tabela_total.groupby('Produto').sum()
tabela_produtos = tabela_produtos[["Quantidade Vendida","Preco Unitario"]].sort_values(by='Quantidade Vendida', ascending=False)
display(tabela_produtos)

#Obs: oque cada tag faz? a groupby pega o produto especificado na tabela; a sum= soma; a sort_values: ordena os valores; o by é o produto em questão, e o ascending é a ordem que é mostrada
```

```python
# Calcular o produto que mais faturou

tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']


tabela_faturamento = tabela_total.groupby('Produto').sum()
display(tabela_faturamento)


```

```python
#Calcular a loja que mais vendeu - Criar um Dashboard

tabela_lojas = tabela_total.groupby('Loja').sum()
tabela_lojas = tabela_lojas[['faturamento']].sort_values(by='faturamento', ascending=False)
display(tabela_lojas)
```

```python
#Juntamente com a biblioteca plotly.express criarei o grafico

grafico = px.bar(tabela_lojas, x=tabela_lojas.index , y='faturamento')
grafico.show()
```
