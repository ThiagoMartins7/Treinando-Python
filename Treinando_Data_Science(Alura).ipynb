```python


import pandas as pd

uri = 'https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula5.1/movies.csv' #Onde os dados estão armazenados
filmes = pd.read_csv(uri) #DataFrame do pandas
filmes.head() #dentro do () você coloca o numero do linhas para ele mostrar

```

```python
uri = 'https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula1.2/ratings.csv' #Onde os dados estão armazenados
notas = pd.read_csv(uri) #DataFrame do pandas
notas.head() #dentro do () você coloca o numero do linhas para ele mostrar
```

```python
notas.columns = ['usuarioId','filmeId','nota','momento'] #Mudar o nome das colunas
notas.head() #dentro do () você coloca o numero do linhas para ele mostrar
```

```python
notas['nota'].head() #é chamado de Series por ser uma sequencia de numeros
```

```python
notas['nota'].unique() #Unique é pra mostrar os valores unicos
```

```python
notas['nota'].mean() #Mean significa Media
```
